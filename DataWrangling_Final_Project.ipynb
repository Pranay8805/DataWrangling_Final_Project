5. Data Cleaning
 a. Remove all duplicates observations
 b. Find missing values in all columns
 c. Remove all observations where continent column value is 
missing
 Tip : using subset parameter in dropna
 d. Fill all missing values with 0



```python
#a. Remove all duplicates observations

data=data.drop_duplicates()
data.reset_index(inplace=True, drop=True)
data
```

```python
#Checking for Null values
data.isna().sum()
```

```python
data=data[data['iso_code'].notna()]
data.reset_index(inplace=True, drop=True)
data

data=data[data['continent'].notna()]
data.reset_index(inplace=True, drop=True)
data

data=data[data['new_cases_smoothed'].notna()]
data.reset_index(inplace=True, drop=True)
data

data=data[data['new_deaths_smoothed'].notna()]
data.reset_index(inplace=True, drop=True)
data

data['total_deaths']=data['total_deaths'].fillna(data['total_deaths'].mean())
data['total_cases']=data['total_cases'].fillna(data['total_cases'].mean())
data['new_cases']=data['new_cases'].fillna(data['new_cases'].mean())
data['female_smokers']=data['female_smokers'].fillna("None")
data['male_smokers']=data['male_smokers'].fillna("None")

data.reset_index(inplace=True, drop=True)


data=data.drop(['extreme_poverty'],axis=1)
data
```

```python
#Checking for Null values
data.isna().sum()
```

```python
# drop duplicates again

data=data.drop_duplicates()
data.reset_index(inplace=True, drop=True)
data
```
